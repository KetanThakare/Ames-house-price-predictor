```python
from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.

        Columns of dtype object are imputed with the most frequent value 
        in column.

        Columns of other types are imputed with mean of column.

        """
    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)

ah_data = DataFrameImputer().fit_transform(ah_data)
ah_data.sample(5)   
```

```python
#one hot encoding the categorical variables
ah_dataOHE = pd.get_dummies(ah_data)
ah_dataOHE.sample(5)
#ah_dataOHE.info()
```

```python
#get location of SalePrice in dataset
#ah_dataOHE.columns.get_loc("SalePrice")
x = ah_dataOHE.loc[:, ah_dataOHE.columns != "SalePrice"]
x = x.loc[:, x.columns!= "Id"]
y = ah_dataOHE.loc[:, 'SalePrice']
#print(x.sample(5))
#print(y.sample(5))
ah_dataOHE.dtypes
print(x.head(5))
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(x, y)
feature_scores = model.feature_importances_
print(model.feature_importances_)
```
